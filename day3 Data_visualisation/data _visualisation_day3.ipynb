# Data visualisation

**Matplotlib**

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
stu=pd.read_csv('StudentsPerformance.csv')
```

```python
stu.head()
```

```python
stu.describe()
```

***histogram***

```python
stu.hist()
```

***bar graph***

```python
stu['race/ethnicity'].value_counts().plot(kind="bar");
```

# Seaborn



***countplot***

```python
sns.countplot(x=stu['gender'])
plt.title('Number of males and females in the dataset')
```

```python
sns.countplot(x=stu['race/ethnicity'])
plt.title('Number of different races people')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='parental level of education', data = stu, hue='gender')
plt.show()
```

```python
plt.figure(figsize=(12,8))
p = sns.countplot(x="math score", data = stu, palette="muted")
```

***bar plot***

```python
sns.barplot(x='gender',y='math score',data=stu)
```

***box plot***

```python
title="Compare the math score"

sns.boxplot(x="gender", y="math score", data=stu)

# increasing font size
plt.title(title, fontsize=26)
# Show the plot
plt.show()
```

***scatter plot***

```python
title="math score"
sns.stripplot(x="gender", y="math score",data=stu) 
plt.title(title, fontsize=16)
```

```python
title="math score"
sns.swarmplot(x="gender", y="math score",data=stu) 
plt.title(title, fontsize=16)
```

***point plot***

```python
title="math score"
sns.pointplot(x="math score", y="race/ethnicity",hue="gender",data=stu,palette={"male":"g", "female":"m"},markers=["^","o"],linestyles=["-","--"]) 
plt.title(title,fontsize=16)
```

***heatmap***

```python
plt.figure(figsize=(7,4)) 
sns.heatmap(stu.corr(),annot=True,cmap='cubehelix_r') #draws  heatmap with input as the correlation matrix calculted by(iris.corr())
plt.show()
```

***violin plot***

```python
sns.violinplot(x="math score", y="gender",data=stu)
```

***Factor plot***

```python
sns.factorplot(x="lunch", hue="gender", col="race/ethnicity", data=stu, kind="count", col_wrap=4, sharey=False);
```
